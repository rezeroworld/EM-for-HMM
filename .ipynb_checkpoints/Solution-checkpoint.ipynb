## Introduction

The file `EMGaussian.train` contains samples of data $\{x_t\}_{t=1}^T$ where $x_t \in \mathbb{R}^2$, with one datapoint per row. `EMGaussian.test` is structured similarly. This is the same data we used in Homework 3, but this time we use an HMM model to account for the possible temporal structure of the data. This means that we now consider each row of the dataset to be a point $x_t \in \mathbb{R}^2$ corresponding to some temporal process, rather than thinking of them as *independent* samples as we did in the last homework. 

We consider the following HMM model: the chain $(z_t)_{t=1}^T$ has  $K=4$ possible states, with an initial probability distribution $\pi\in\Delta_4$ and a probability transition matrix  $A \in \mathbb{R}^{4 \times 4}$ where $A_{ij} = p(z_t=i | z_{t-1} = j),$ and conditionally on the current state $z_t$, we have observations obtained from Gaussian emission probabilities $x_t| (z_t=k) \sim \mathcal{N}(x_t | \mu_k, \Sigma_k)$.  This is thus a generalization of a GMM since we now allow for time dependencie across the latent states $z_t$.

This exercise has several implementation objectives:
* **Sum-product**: probabilistic inference on the HMM
* **Expectation-Maximization**: parameter estimation
* **Viterbi**: decoding.

## Sum-product [15 pts]

```python
# Just a helper function
def log_multivariate_gaussian(X, mean, cov):
    part1 = 1 / ( ((2* np.pi)**(len(mean)/2)) * (np.linalg.det(cov)**(1/2)) )
    part2 = (-1/2) * ((X-mean).dot(np.linalg.inv(cov))).dot((X-mean).T)
    return np.log(part1) + np.diag(part2)
```

### $\alpha$, $\beta$ recursions

Implement the $\alpha$ and $\beta$-recursions seen in class (and that can be found in chapter 12 of Mike's book with slightly different notation). Recall that $\alpha(z_t) := p(z_t, x_{1:t})$, $\beta(z_t) := p(x_{(t+1):T} | z_t)$. Implement also a function to compute the emission probabilties $\epsilon_k(x_t) := p(x_t|z_t=k) = \mathcal{N}(x_t|\mu_k, \Sigma_k)$.

For numerical stability reasons, you are expected to implement your algorithms using **log probabilities** unless noted explicitly!



```python
def log_emission(X, mus, sigmas) :
    """
    (Log) probabilities under a Gaussian emission model for a time-homogeneous HMM 

    Inputs:
        X: [Tx2] matrix of observations
        mus: [Kx2] matrix of latent-conditional emission means
        sigmas: [Kx2x2] tensor of latent-conditional emission covariance matrices
         
    Returns:
        log_eps: [TxK] matrix of log emission probabilities: log p(x_t | z_t = k)
    """
    T, K = X.shape[0], mus.shape[0]

    # TODO
    log_eps = np.zeros((T, K))
    for k in range(K):
        log_eps[:,k] = log_multivariate_gaussian(X, mus[k], sigmas[k])

    return log_eps

def log_alpha_recursion(X, A, log_eps, pi) :
    """
    (Log) alpha recursion for a time-homogeneous HMM with Gaussian emissions

    Inputs:
        X: [Tx2] matrix of observations
        A: [KxK] transition matrix
        log_eps: [TxK] matrix of log emission probabilities: log p(x_t | z_t = k)
        pi: [Kx1] initial latent state distribution
    
    Returns:
        log_alpha: [TxK] vector containing log p(z_t , x_{1:t})
    """
    T, K = log_eps.shape

    # TODO
    log_alpha = np.zeros((T, K))
    log_alpha[0,:] = log_eps[0,:] + np.log(pi)

    for i in range(1, T):
        for j in range(K):
            a = np.log(A[j,:]) + log_alpha[i-1,:]
            a_ = np.max(a)
            a_f = a - a_
            log_alpha[i,j] = log_eps[i, j] + a_ + np.log(np.sum(np.exp(a_f)))

    return log_alpha

def log_beta_recursion(X, A, log_eps) :
    """
    (Log) beta recursion for a time-homogeneous HMM with Gaussian emissions

    Inputs:
        X: [Tx2] matrix of observations
        A: [KxK] transition matrix
        log_eps: [TxK] matrix of log emission probabilities: log p(x_t | z_t = k)
    
    Returns:
        log_beta: [TxK] vector containing log p(z_{t+1:T} | z_t)
    """
    T, K = log_eps.shape
    
    # TODO
    log_beta = np.zeros((T, K))  
    for i in range(T-1, 0, -1):
        for j in range(K):
            a = np.log(A[:,j]) + log_eps[i,:] + log_beta[i,:] 
            a_ = np.max(a)
            a_f = a - a_
            log_beta[i-1, j] = a_ + np.log(np.sum(np.exp(a_f)))

    # This redundant line is just a reminder to leave the T-th row (python's (T-1)-th)
    # empty: Look at the equation for beta(z_T). We do this to keep equal sizes
    # for alpha and beta recursions.
    log_beta[T-1] = np.zeros((K,))

    return log_beta
```

Using the two functions above, implement the computation of the *smoothing* distribution $p(z_t|x_1,\dots,x_T)$ and pair-marginals $p(z_t,z_{t+1}|x_1,\dots,x_T)$. Here use log probabilities from the $\alpha$ and $\beta$, but return a normal (not log!) probability, i.e, a number in $[0,1]$.

```python
def smoothing(log_alpha, log_beta):
    """
    Smoothing probabilities for a time-homogeneous HMM with Gaussian emissions

    Inputs:
        log_alpha: [TxK] matrix containing log p(z_t , x_{1:t})
        log_beta: [TxK] matrix containing log p(z_{t+1:T} | z_t)
    
    Returns:
        gamma: [TxK] matrix of smoothing probabilities p(z_t | x_{1:T})
    """
    T, K = log_alpha.shape

    # TODO
    gamma = np.zeros((T, K))

    log_gamma = log_alpha + log_beta
    log_denominator = np.empty(shape=(log_gamma.shape[0]))
        
    for i in range(T):
        a = log_alpha[i,:] + log_beta[i,:]
        a_ = np.max(a)
        a_f = a - a_
        log_denominator[i] = a_ + np.log(np.sum(np.exp(a_f)))

    return np.exp(log_gamma - log_denominator[:, np.newaxis])

def pair_marginals(log_alpha, log_beta, A, log_eps):
    """
    Pair marginals for a time-homogeneous HMM with Gaussian emissions

    Inputs:
        log_alpha: [TxK] matrix containing log p(z_t , x_{1:t})
        log_beta: [TxK] matrix containing log p(z_{t+1:T} | z_t)
        A: [KxK] transition matrix
        log_eps: [TxK] matrix of log emission probabilities: log p(x_t | z_t = k)
    
    Returns:
        psi: [TxKxK] numpy tensor of pair marginal probabilities p(z_t, z_{t+1} | x_{1:T})
    """
    T, K = log_alpha.shape

    # TODO
    psi = np.zeros((T, K, K))

    for i in range(T-1):
        log_nominator = log_alpha[i,:] + log_beta[i+1,:] + log_eps[i+1,:] + np.log(A) 
        for j in range(K):
            a = log_nominator[j,:]
            a_ = np.max(a)
            a_f = a - a_
            log_denominator = a_ + np.log(np.sum(np.exp(a_f)))
            psi[i,j,:] = np.exp(log_nominator[j,:] - log_denominator)

    # Just as above, we keep psi of length T on the first dimension
    psi[T-1, ...] = np.zeros((K,K))

    return psi
```

## Expectation-Maximization [20 pts]

```python
# Some helper functions
def update_E_step(X, mus, sigmas, A, pi):
    log_eps = log_emission(X, mus, sigmas)
    log_alpha = log_alpha_recursion(X, A, log_eps, pi)
    log_beta = log_beta_recursion(X, A, log_eps)
    tau = smoothing(log_alpha, log_beta)
    psi = pair_marginals(log_alpha, log_beta, A, log_eps)
    return tau, psi

def update_mus_sigmas(X_tr, tau):
    K = tau.shape[1]
    for j in range(K):
        weight = tau[:, j]
        total_weight = weight.sum()
        mus[j] = (X_tr * weight[:,np.newaxis]).sum(axis=0) / total_weight

        s = (weight[:,np.newaxis] * (X_tr - mus[j])).T.dot(X_tr - mus[j])
        sigmas[j] = s / total_weight
    return mus, sigmas

def update_A(psi):
    K = psi.shape[1]
    for i in range(K):
        for j in range(K):
            nominator = np.sum(psi[1:,i,j])
            denominator = np.sum(psi[:, :, j].flatten())
            A[i,j] = nominator / denominator
    return A

def update_M_step(X_tr, tau, psi):
    pi = tau.sum(axis=0)
    mus, sigmas = update_mus_sigmas(X_tr, tau)
    A = update_A(psi)
    return pi, mus, sigmas, A

def average_llh(X, mus, sigmas, A, pi):
    T = X.shape[0]
    log_eps = log_emission(X, mus, sigmas) 
    log_alpha = log_alpha_recursion(X, A, log_eps, pi)
    return logsumexp(log_alpha[-1], axis=0) / T
```

### M-step updates
Derive the M-step update for $\hat{\pi}$, $\hat{A}$, $\hat{\mu}_k$ and $\hat{\Sigma}_k$ (for $k=1,\ldots, 4)$ during the EM algorithm, as a function of the quantities computed during the E step. Note that for the estimate of $\pi$, we only have *one* long chain here!

**Answer:**
Because we have a mixture model here it means that in the M step we will be using the expected complete log likelihood as we did for the GMM, the rule is the following:
$E_q[log(z,x|\theta)] = E[\sum_{i=1}^N log(p(z_1^{(i)})\prod_{t=1}^{T_i}p(\bar{x}_t^{(i)}|z_t^{(i)})\prod_{t=2}^{T_i}p(z_t^{(i)}|z_{t-1}^{(i)}))]$ 
$E_q[log(z,x|\theta)] = E[\sum_{i=1}^N log(p(z_1^{(i)})) + log(\sum_{t=1}^{T_i}p(\bar{x}_t^{(i)}|z_t^{(i)}))+ log(\sum_{t=2}^{T_i}p(z_t^{(i)}|z_{t-1}^{(i)}))]$
$E_q[log(z,x|\theta)] = E[\sum_{i=1}^N \sum_k z_{1,k}^{(i)}log(\pi_k) + \sum_{t=1}^{T_i}\sum_k z_{t,k}^{(i)}log(N(\bar{x}_t^{(i)}|\mu_k, \Sigma_k))+ \sum_{t=2}^{T_i}\sum_{l,m}z_{t,l}^{(i)}z_{t-1,m}^{(i)}log(A_{l,m})]$
$E_q[log(z,x|\theta)] = \sum_{i=1}^N \sum_k E[z_{1,k}^{(i)}]log(\pi_k) + \sum_{t=1}^{T_i}\sum_k E[z_{t,k}^{(i)}]log(N(\bar{x}_t^{(i)}|\mu_k, \Sigma_k))+ \sum_{t=2}^{T_i}\sum_{l,m}E[z_{t,l}^{(i)}z_{t-1,m}^{(i)}]log(A_{l,m})]$
$E[z_{t,k}^{(i)}] = q(z_{t,k}^{(i)} = 1) = \tau_{t,k}^{(i)}$

$E[z_{t,l}^{(i)}z_{t-1,m}^{(i)}] = q(z_{t,l}^{(i)} = 1, z_{t-1,m}^{(i)} = 1) = p(z_{t,l}^{(i)} = 1, z_{t-1,m}^{(i)} = 1|\bar{x}_{1:T_i}, \theta) = \Psi_{t,l,m}^{(i)}$

Let's compute the maximizer for each quantity:

$\hat{\pi}_k = argmax_{\pi_k} E_q[log(z,x|\theta)]$, by taking the derivative which only involves the terms where $\pi_k$ is present, we get the same term as we did in GMM (there is no t included) which we can resolve by using lagrange multipliers like we did in the previous homework. at the end we get: $\hat{\pi}_k = \frac{\sum_i\tau_{1,k}^{(i)}}{N}$

For $\hat{\mu_k}$ and $\hat{\Sigma_k}$ only the second term of the expectation will be used, by deriving, setting equal to $0$ and resolving a simple equation we get:

$\hat{\mu_k} = \frac{\sum_{i=1}^{N}\sum_{t=1}^{T_i} \tau_{t,k}^{(i)}x_t^{(i)}}{\sum_{i=1}^{N}\sum_{t=1}^{T_i} \tau_{t,k}^{(i)}}$, $\hat{\Sigma_k} = \frac{\sum_{i=1}^{N}\sum_{t=1}^{T_i} \tau_{t,k}^{(i)}(x_t^{(i)} - \mu_k)^T(x_t^{(i)} - \mu_k)}{\sum_{i=1}^{N}\sum_{t=1}^{T_i} \tau_{t,k}^{(i)}}$

For $\hat{A}_{l,m}$ only the third term will be involved in the derivative computation:

$\frac{d}{dA_{l,m}}E_q[log(z,x|\theta)] = \frac{d}{dA_{l,m}} \sum_{i=1}^{N}\sum_{t=2}^{T_i}\sum_{l,m}E[z_{t,l}^{(i)}z_{t-1,m}^{(i)}]log(A_{l,m})] = \frac{d}{dA_{l,m}} \sum_{i=1}^{N}\sum_{t=2}^{T_i}\sum_{l,m}\Psi^{(i)}_{t,l,m} log(A_{l,m})  = \sum_{i=1}^{N}\sum_{t=2}^{T_i}\frac{\Psi^{(i)}_{t,l,m}}{A_{l,m}}$

By setting it equal to $0$ we can resolve the equation using lagrande multipliers and get:

$\hat{A}_{l,m} = \frac{\sum_{i=1}^{N}\sum_{t=2}^{T_i} \Psi^{(i)}_{t,l,m}}{\sum_{u}\sum_{i=1}^{N}\sum_{t=2}^{T_i}\Psi^{(i)}_{t,u,m}}$

### EM implementation

Implement the EM algorithm to learn the parameters of the model ($\pi,A,\mu_k,\Sigma_k,\: k=1\ldots,4$). Use the parameters from question 2 for initialization.

```python
def exp_max(X_tr, X_ts, init_params):
    """
    Estimates the parameters of an HMM using training data X via the EM algorithm

    Inputs:
        X_tr: [T_trainx2] matrix of training observations
        X_tr: [T_testx2] matrix of test observations
        init_params: tuple of initialization parameters from previous question
        
    Returns:
        pi: [K] estimated initial latent distribution
        A: [KxK] estimated transition matrix
        mus: [Kx2] matrix of estimated emission means
        sigmas: [Kx2x2] tensor of estimated emission covariance matrices
        train_avg_llike: list containing the average training log likelihood on each iteration
        test_avg_llike: list containing the average test log likelihood on each iteration
    """

    # Set initialization from parameters given in previous question 
    (K, pi, A, mus, sigmas) = init_params

    # TODO
    train_avg_llike  = []
    test_avg_llike  = []
    
    for _ in range(6):
        # E step: compute tau: [TxK] which is the smoothing distribution
        tau, psi = update_E_step(X_tr, mus, sigmas, A, pi)
        
        # M step: Compute mus, sigmas, pi, A
        pi, mus, sigmas, A = update_M_step(X_tr, tau, psi)

        # compute the average train llk
        train_llike = average_llh(X_tr, mus, sigmas, A, pi)
        train_avg_llike.append(train_llike)

        # compute the average test llk
        test_llike = average_llh(X_ts, mus, sigmas, A, pi)
        test_avg_llike.append(test_llike)
          
    return pi, A, mus, sigmas, train_avg_llike, test_avg_llike
```

### Viterbi implementation

Implement Viterbi decoding.

```python
def viterbi(X, pi, A, mus, sigmas):
    """
    Viterbi decoding for a time-homogeneous HMM with Gaussian emissions

    Inputs:
        X: [Tx2] matrix of observations
        pi: [Kx1] initial latent state distribution
        A: [KxK] transition matrix
        mus: [Kx2] matrix of latent-conditional emission means
        sigmas: [Kx2x2] tensor of latent-conditional emission covariance matrices
         
    Returns:
        z: [TxK] one-hot encoding of most probable state sequence z_{1:T} given x_{1:T}
    """
    
    T, K = X.shape[0], mus.shape[0]

    # TODO
    z = np.zeros((T, K))
    w = np.zeros((T, K)) 
    z_ = np.zeros((T, K))

    log_eps = log_emission(X, mus, sigmas)

    for j in range(K):
        z_[0, j] = np.log(pi[j]) + log_eps[0,j]

    for t in range(1, T):
        p = z_[t-1] + np.log(A)
        z_[t] = log_eps[t] + np.max(p, axis=1)
        w[t - 1] = np.argmax(p, axis=1)

    z_f = np.argmax(z_[T-1])
    z[T-1, z_f] = 1.
    for t in range(T-2, -1, -1):
        z_f = int(w[t, z_f])
        z[t, z_f] = 1.

    return z
```
